```python
import numpy as np
import matplotlib.pyplot as plt
types = ['FP32', 'FP16', 'FX32', 'FX24', 'FX16']
for datatype in types:
    fidelity_file = f'./fidelity/qft/qft_{datatype}_Fidelity.txt'
    fidelity_data = np.loadtxt(fidelity_file)
    if datatype in ['FP32', 'FX32']:
        print(fidelity_data)
    plt.plot(range(3, 18), fidelity_data, label=datatype)
plt.ylim(0, 1.1)
plt.xticks(range(3, 18))
plt.xlabel('# qubits')
plt.ylabel('Fidelity')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
types = ['FP32', 'FX32', 'FX24']
num_qubits = 17
depths = 10
num_repeat = 100

for datatype in types:
    fidelitiesss = np.zeros((num_qubits - 2, depths, num_repeat))
    for i in range(1, num_repeat + 1):
        for depth in range(1, depths + 1):
            fidelity_file = f'./fidelity/qrc/{datatype}/{i}/qrc__{depth}depth_Fidelity.txt'
            fidelity_data = np.loadtxt(fidelity_file)
            fidelitiesss[:, depth - 1, i - 1] = fidelity_data

    # get mean of 3-d fidelitiesss

    mean_fidelitiesss = np.mean(fidelitiesss)
    print(mean_fidelitiesss)
```
