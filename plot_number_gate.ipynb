```python
import numpy as np

import matplotlib.pyplot as plt

# Create x and y values
x = np.arange(1, 11)
y = np.arange(1, 11)

# Create a meshgrid of x and y values
X, Y = np.meshgrid(x, y)

# Create heat values
heat_values = np.random.randint(0, 11, size=(10, 10))

# Plot the heatmap
plt.imshow(heat_values, cmap='hot', origin='lower', extent=[1, 10, 1, 10])
plt.colorbar()

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Heatmap Example')

# Show the plot
plt.show()

```
